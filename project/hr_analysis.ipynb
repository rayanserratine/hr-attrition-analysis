```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df['attrition'] = df['attrition'].map({'Yes': 1, 'No': 0})
```

```python
drop_cols = [
    'employeecount',
    'over18',
    'standardhours'
]

df = df.drop(columns=drop_cols)
```

```python
df['salary_level'] = pd.cut(
    df['monthlyincome'],
    bins=[0, 3000, 7000, 20000],
    labels=['Low', 'Medium', 'High']
)
```

```python
df.head()
```

```python
df['tenure_group'] = pd.cut(
    df['yearsatcompany'],
    bins=[0, 3, 7, 40],
    labels=['New', 'Mid', 'Senior']
)
```

```python
df['engagement_score'] = (
    df['jobsatisfaction'] +
    df['worklifebalance'] +
    df['jobinvolvement']
)
```
